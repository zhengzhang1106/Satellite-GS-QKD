```python
NUM_DATA = 50
amps = [0.0, 0.5, 1.0]
#demand_list = [0.05 + 0.004*i for i in range(NUM_DATA)]
demand_list = [0.01 + 0.02*i for i in range(NUM_DATA)]

fig, axes = plt.subplots(2, 2, figsize=(8, 6), sharey=True)
axes = axes.flatten()

time_hours = np.array(syst.T) * (24.0 / len(syst.T))

# ---- First 3 panels (noise = 0) ----
for idx_ax, amp in enumerate(amps):
    ax = axes[idx_ax]
    all_demands = []

    for mean in demand_list:
        d = generate_demands(
            gnodes,
            syst,
            mean_kbps=mean,
            amp=amp,
            noise_std=0,
            pattern="stadium"
        )[0]

        all_demands.append(d.K_REQ)

    all_demands = np.array(all_demands)

    d_min = all_demands.min(axis=0)
    d_med = np.median(all_demands, axis=0)
    d_max = all_demands.max(axis=0)

    ax.fill_between(time_hours, d_min, d_max, alpha=0.3)
    ax.plot(time_hours, d_med, linewidth=2)

    ax.set_title(f"amp = {amp}, noise = 0", fontsize=13)
    ax.set_xlim(0, 24)
    ax.set_xticks([0, 6, 12, 18, 24])
    ax.set_xticklabels(["12AM", "6AM", "12PM", "6PM", "12AM"], fontsize=13)
    ax.grid(True, linestyle="--", alpha=0.6)


# ---- 4th panel (amp = 0.5 with noise) ----
ax = axes[3]
all_demands = []

for mean in demand_list:
    d = generate_demands(
        gnodes,
        syst,
        mean_kbps=mean,
        amp=0.5,
        noise_std=0.2,   # <-- noisy case
        pattern="stadium"
    )[0]

    all_demands.append(d.K_REQ)

all_demands = np.array(all_demands)

d_min = all_demands.min(axis=0)
d_med = np.median(all_demands, axis=0)
d_max = all_demands.max(axis=0)

ax.fill_between(time_hours, d_min, d_max, alpha=0.3)
ax.plot(time_hours, d_med, linewidth=2)

ax.set_title("amp = 0.5, noise > 0", fontsize=13)
ax.set_xlim(0, 24)
ax.set_xticks([0, 6, 12, 18, 24])
ax.set_xticklabels(["12AM", "6AM", "12PM", "6PM", "12AM"], fontsize=13)
ax.grid(True, linestyle="--", alpha=0.6)

# ---- Labels ----
axes[0].set_ylabel("SKR Demand (kbps)", fontsize=13)
axes[2].set_ylabel("SKR Demand (kbps)", fontsize=13)

for ax in axes:
    ax.set_xlabel("Time of the Day", fontsize=13)

plt.tight_layout()
plt.savefig("demand_envelope_2x2.pdf", format='pdf', transparent=True, bbox_inches='tight', pad_inches=0.01)
plt.show()
```
