## QKP Analysis Experiment

```python
####################################################################################################
########################################### MAIN EXPERIMENT ########################################
####################################################################################################

for (amp, noise) in amps_noise:

    print(f"\nRunning experiments for amp = {amp}, noise={noise}")

    for idx, mean in enumerate(demand_list):

        num_runs = NUM_NOISE_RUNS if noise != 0 else 1

        grd_runs = []
        grd_qkp_runs = []
        saved_runs = []

        for run in range(num_runs):

            ##################################################################
            ######################## DEMAND GENERATION #######################
            ##################################################################

            demands = [
                generate_demands(
                    gnodes,
                    syst,
                    mean_kbps=mean,
                    amp=amp,
                    noise_std=noise,
                    pattern="stadium"
                )[0]
            ]

            if run == 0:
                key_req_all[(amp,noise)][idx] = (
                    sum(d.K_REQ[t] * syst.THETA for d in demands for t in syst.T) * 1e-6
                )

            ##################################################################
            ######################## GREEDY ROUTING ##########################
            ##################################################################

            Z = [[[0 for t in syst.T] for d in demands] for l in links]

            ordr_dmnds = []
            for t in syst.T:
                ordr_dmnds.append(
                    sorted(demands, key=lambda d: d.K_REQ[t], reverse=True)
                )

            for t in syst.T:

                max_cnct_up  = [1] * len(nodes)
                max_cnct_dwn = [1] * len(nodes)

                for idx_d, d in enumerate(ordr_dmnds[t]):

                    src, dst = d.n1, d.n2
                    l1_sel, l2_sel = None, None
                    max_skr = 0

                    for l_1 in links:
                        for l_2 in links:
                            if (
                                l_1.n1 == src and
                                l_2.n2 == dst and
                                l_1.n2 == l_2.n1
                            ):
                                mid = l_1.n2
                                btln_skr = min(l_1.K_MAX[t], l_2.K_MAX[t])

                                if (
                                    btln_skr > max_skr and
                                    max_cnct_up[nodes.index(src)] > 0 and
                                    max_cnct_dwn[nodes.index(dst)] > 0 and
                                    max_cnct_up[nodes.index(mid)] > 0 and
                                    max_cnct_dwn[nodes.index(mid)] > 0
                                ):
                                    max_skr = btln_skr
                                    l1_sel, l2_sel = l_1, l_2

                    if l1_sel is not None:

                        mid = l1_sel.n2

                        max_cnct_up[nodes.index(src)] = 0
                        max_cnct_dwn[nodes.index(dst)] = 0
                        max_cnct_up[nodes.index(mid)] = 0
                        max_cnct_dwn[nodes.index(mid)] = 0

                        Z[links.index(l1_sel)][idx_d][t] = 1
                        Z[links.index(l2_sel)][idx_d][t] = 1

            ##################################################################
            ######################## LP SOLVES ################################
            ##################################################################

            solution_grd, key_grd, _ = offline_lp(
                gnodes, hnodes, links, demands, 0, 2, Z
            )

            solution_qkp, key_grd_qkp, _ = offline_lp(
                gnodes, hnodes, links, demands, 1, 2, Z
            )

            grd_runs.append(key_grd * 1e-6)
            grd_qkp_runs.append(key_grd_qkp * 1e-6)

            ##################################################################
            ################ MAX SAVED SECRET KEY (FROM a[l,t]) ##############
            ##################################################################

            a_matrix = solution_qkp["a"]  # shape: [link, time]

            saved_per_time = []

            for t_idx in range(len(syst.T)):
                total_t = sum(a_matrix[idx_l, t_idx] for idx_l in range(len(links)))
                saved_per_time.append(total_t)

            max_saved = max(saved_per_time) * 1e-6
            saved_runs.append(max_saved)

        ######################################################################
        ######################## STORE STATISTICS ############################
        ######################################################################

        key_srv_grd_mean[(amp,noise)][idx] = np.mean(grd_runs)
        key_srv_grd_std[(amp,noise)][idx]  = np.std(grd_runs)

        key_srv_grd_qkp_mean[(amp,noise)][idx] = np.mean(grd_qkp_runs)
        key_srv_grd_qkp_std[(amp,noise)][idx]  = np.std(grd_qkp_runs)

        max_saved_bits_mean[(amp,noise)][idx] = np.mean(saved_runs)
        max_saved_bits_std[(amp,noise)][idx]  = np.std(saved_runs)

        print(f"amp={amp}, demand_idx={idx}")
```

## Algorithm Tests

```python
NUM_DATA    = 50
N_RUNS      = 10   # number of Monte-Carlo runs

demand_list = [0.01 + 0.2 * i for i in range(NUM_DATA)]
```

```python
key_srv_grd_qkp = [0 for i in range(NUM_DATA)]
key_srv_rnd_qkp = [0 for i in range(NUM_DATA)]
key_srv_rlx_qkp = [0 for i in range(NUM_DATA)]
key_srv_opt_qkp = [0 for i in range(NUM_DATA)]

a_lst_grd_qkp = [0 for i in range(NUM_DATA)]
a_lst_rnd_qkp = [0 for i in range(NUM_DATA)]
a_lst_rlx_qkp = [0 for i in range(NUM_DATA)]
a_lst_opt_qkp = [0 for i in range(NUM_DATA)]

key_req = [0 for i in range(NUM_DATA)]

for idx, mean in enumerate(demand_list):
    demands = []
    # pairs = [(i, j) for i in range(len(gnodes)) for j in range(i+1, len(gnodes))]

    # for (i, j) in pairs:
    #     gs1, gs2 = gnodes[i], gnodes[j]

    #     k_vals = []
    #     for t in syst.T:
    #         k_vals.append(mean)  # ensure positive rate

    #     demands.append(demand(k_vals, gs1, gs2))

    demands = generate_demands(gnodes, syst, mean_kbps=mean, amp=0.5, noise_std=0, pattern="sinusoidal")

    key_req[idx] = sum(d.K_REQ[t] * syst.THETA * 1e-3 for d in demands for t in syst.T)

    ########################################## Greedy (Scarcity + Contention Aware, 1-hop only) ##########################################

    # Z[l][d][t] = 1 if link l is used by demand d at time t
    Z = [[[0 for t in syst.T] for d in demands] for l in links]
    
    # Stored QKP bits per link (already scaled by THETA)
    qkp_s = [0 for _ in links]
    
    for t in syst.T:
    
        # Per-HAP connection limits (up/down)
        max_cnct_up  = [1] * len(hnodes)
        max_cnct_dwn = [1] * len(hnodes)
    
        # --------------------------------------------------------------------------------------------------
        # Step 1: Enumerate feasible 1-hop paths for each demand
        # A path (l1, l2) is feasible if:
        #   - fresh keys can serve it on both links, OR
        #   - stored QKP can cover the demand on both links
        # --------------------------------------------------------------------------------------------------
        feasible_paths = {}   # d -> list of (idx_l1, idx_l2, mid)
    
        for d in demands:
            src, dst = d.n1, d.n2
            feasible_paths[d] = []
    
            for idx_l1, l1 in enumerate(links):
                for idx_l2, l2 in enumerate(links):
                    if l1.n1 == src and l2.n2 == dst and l1.n2 == l2.n1:
                        mid = l1.n2
    
                        fresh_ok = (
                            l1.K_MAX[t] >= d.K_REQ[t] and
                            l2.K_MAX[t] >= d.K_REQ[t]
                        )
    
                        qkp_ok = (
                            qkp_s[idx_l1] / syst.THETA >= d.K_REQ[t] and
                            qkp_s[idx_l2] / syst.THETA >= d.K_REQ[t]
                        )
    
                        if fresh_ok or qkp_ok:
                            feasible_paths[d].append((idx_l1, idx_l2, mid))
    
        # --------------------------------------------------------------------------------------------------
        # Step 2: Order demands by scarcity (least number of feasible paths first)
        # --------------------------------------------------------------------------------------------------
        ordered_demands = sorted(
            demands,
            key=lambda d: (
                len(feasible_paths[d]),
                -d.K_REQ[t]
            )
        )
    
        # # --------------------------------------------------------------------------------------------------
        # # Step 3: Compute path contention
        # # How many demands can use each (l1, l2) path
        # # --------------------------------------------------------------------------------------------------
        # path_usage = {}   # (idx_l1, idx_l2) -> count
    
        # for d in demands:
        #     for (idx_l1, idx_l2, _) in feasible_paths[d]:
        #         key = (idx_l1, idx_l2)
        #         path_usage[key] = path_usage.get(key, 0) + 1
        
        # --------------------------------------------------------------------------------------------------
        # Step 3 (FIXED): Compute link contention
        # How many feasible paths (across all demands) include each link
        # --------------------------------------------------------------------------------------------------
        link_usage = [0] * len(links)
        
        for d in demands:
            for (idx_l1, idx_l2, _) in feasible_paths[d]:
                link_usage[idx_l1] += 1
                link_usage[idx_l2] += 1
    
        # --------------------------------------------------------------------------------------------------
        # Step 4: Greedy allocation
        # For each demand, try least-contended feasible paths first
        # --------------------------------------------------------------------------------------------------
        for d in ordered_demands:
    
            l1_sel = None
            l2_sel = None
            used_qkp_only = False
    
            # # Order this demand's paths by least contention
            # paths = sorted(
            #     feasible_paths[d],
            #     key=lambda p: path_usage[(p[0], p[1])]
            # )
            
            # Order this demand's paths by least link contention
            paths = sorted(
                feasible_paths[d],
                key=lambda p: link_usage[p[0]] + link_usage[p[1]]
            )
    
            for idx_l1, idx_l2, mid in paths:
    
                # Respect HAP connection limits
                if max_cnct_up[hnodes.index(mid)] == 0:
                    continue
                if max_cnct_dwn[hnodes.index(mid)] == 0:
                    continue
    
                # Case 1: Fully served by stored QKP
                if (
                    qkp_s[idx_l1] / syst.THETA >= d.K_REQ[t] and
                    qkp_s[idx_l2] / syst.THETA >= d.K_REQ[t]
                ):
                    qkp_s[idx_l1] -= d.K_REQ[t] * syst.THETA
                    qkp_s[idx_l2] -= d.K_REQ[t] * syst.THETA
                    used_qkp_only = True
                    break
    
                # Case 2: Use fresh generation (possibly mixed with QKP later)
                l1_sel = links[idx_l1]
                l2_sel = links[idx_l2]
                break
    
            # --------------------------------------------------------------------------------------------------
            # Step 5: Update state after allocation
            # --------------------------------------------------------------------------------------------------
            if used_qkp_only:
                continue
    
            if l1_sel is not None:
                mid = l1_sel.n2
                max_cnct_up[hnodes.index(mid)] = 0
    
                l_idx = links.index(l1_sel)
                Z[l_idx][demands.index(d)][t] = 1
                qkp_s[l_idx] += (l1_sel.K_MAX[t] - d.K_REQ[t]) * syst.THETA
    
            if l2_sel is not None:
                mid = l2_sel.n1
                max_cnct_dwn[hnodes.index(mid)] = 0
    
                l_idx = links.index(l2_sel)
                Z[l_idx][demands.index(d)][t] = 1
                qkp_s[l_idx] += (l2_sel.K_MAX[t] - d.K_REQ[t]) * syst.THETA
    
    # --------------------------------------------------------------------------------------------------
    # Final LP refinement
    # --------------------------------------------------------------------------------------------------
    solution, key_srv_grd_qkp[idx], a_lst_grd_qkp[idx] = offline_lp(
        gnodes, hnodes, links, demands, 1, 2, Z
    )
    
    key_srv_grd_qkp[idx] *= 1e-6
    a_lst_grd_qkp[idx]   *= 1e-6
    ######################################################################################################

    # ########################################## Random #######################################################
    # Z = [[[0 for t in syst.T] for d in demands] for l in links]

    # for t in syst.T:
    #     max_cnct_up  = [1 for _ in hnodes]
    #     max_cnct_dwn = [1 for _ in hnodes]
    
    #     shuffled_demands = random.sample(demands, len(demands))
    
    #     for idx_d, d in enumerate(shuffled_demands):
    #         src, dst = d.n1, d.n2
    
    #         feasible_pairs = []
    
    #         for l1 in links:
    #             for l2 in links:
    #                 if (
    #                     l1.n1 == src and
    #                     l2.n2 == dst and
    #                     l1.n2 == l2.n1
    #                 ):
    #                     mid = l1.n2
    #                     if (
    #                         max_cnct_up[hnodes.index(mid)] > 0 and
    #                         max_cnct_dwn[hnodes.index(mid)] > 0 and
    #                         l1.K_MAX[t] > 0 and
    #                         l2.K_MAX[t] > 0
    #                     ):
    #                         feasible_pairs.append((l1, l2))
    
    #         if feasible_pairs:
    #             l1_sel, l2_sel = random.choice(feasible_pairs)
    #             mid = l1_sel.n2
    
    #             Z[links.index(l1_sel)][demands.index(d)][t] = 1
    #             Z[links.index(l2_sel)][demands.index(d)][t] = 1

    #             #print(f"({links.index(l1_sel)}, {idx_d}, {t})")
    #             #print(f"({links.index(l2_sel)}, {idx_d}, {t})")
    
    #             max_cnct_up[hnodes.index(mid)]  = 0
    #             max_cnct_dwn[hnodes.index(mid)] = 0
    
    # solution, key_srv_rnd_qkp[idx], a_lst_rnd_qkp[idx] = offline_lp(gnodes, hnodes, links, demands, 1, 2, Z)

    # key_srv_rnd_qkp[idx] *= 1e-6
    # a_lst_rnd_qkp[idx]   *= 1e-6
    # #########################################################################################################

    # Z = [[[0 for t in syst.T] for d in demands] for l in links]

    # for t in syst.T:
    #     max_up  = [1]*len(nodes)
    #     max_dwn = [1]*len(nodes)
    
    #     shuffled_links = random.sample(list(enumerate(links)), len(links))
    
    #     for idx_l, l in shuffled_links:
    #         u = l.n1
    #         v = l.n2
    
    #         if max_up[node_id[u]] and max_dwn[node_id[v]]:
    #             Z[idx_l][0][t] = 1   # demand index irrelevant, see note below
    
    #             max_up[node_id[u]] = 0
    #             max_dwn[node_id[v]] = 0

    ########################################## Relaxed #######################################################
    solution, key_srv_rlx_qkp[idx], a_lst_rlx_qkp[idx] = offline_relaxed(gnodes, hnodes, links, demands, 1, 2)

    key_srv_rlx_qkp[idx] *= 1e-6
    a_lst_rlx_qkp[idx]   *= 1e-6

    ########################################## Optimal #######################################################
    solution, key_srv_opt_qkp[idx], a_lst_opt_qkp[idx] = offline(gnodes, hnodes, links, demands, 1, 2)

    key_srv_opt_qkp[idx] *= 1e-6
    a_lst_opt_qkp[idx]   *= 1e-6

    print(f"idx: {idx}")
```

```python
# Store statistics
key_srv_rnd_qkp_avg   = [0.0 for _ in range(NUM_DATA)]
key_srv_rnd_qkp_min   = [0.0 for _ in range(NUM_DATA)]
key_srv_rnd_qkp_max   = [0.0 for _ in range(NUM_DATA)]
a_lst_rnd_qkp_avg     = [0.0 for _ in range(NUM_DATA)]
a_lst_rnd_qkp_min     = [0.0 for _ in range(NUM_DATA)]
a_lst_rnd_qkp_max     = [0.0 for _ in range(NUM_DATA)]
key_req               = [0.0 for _ in range(NUM_DATA)]

for idx, mean in enumerate(demand_list):

    # Generate traffic once per demand point
    demands = generate_demands(
        gnodes,
        syst,
        mean_kbps=mean,
        amp=0.5,
        noise_std=0,
        pattern="sinusoidal",
    )

    key_req[idx] = sum(
        d.K_REQ[t] * syst.THETA * 1e-3
        for d in demands
        for t in syst.T
    )

    rnd_results_k_srv = []
    rnd_results_a_lst = []

    ########################################## Random (Monte Carlo) ########################################
    for run in range(N_RUNS):

        Z = [[[0 for t in syst.T] for d in demands] for l in links]

        for t in syst.T:
            max_cnct_up  = [1 for _ in hnodes]
            max_cnct_dwn = [1 for _ in hnodes]

            shuffled_demands = random.sample(demands, len(demands))

            for d in shuffled_demands:
                src, dst = d.n1, d.n2
                feasible_pairs = []

                for l1 in links:
                    for l2 in links:
                        if (
                            l1.n1 == src and
                            l2.n2 == dst and
                            l1.n2 == l2.n1
                        ):
                            mid = l1.n2
                            if (
                                max_cnct_up[hnodes.index(mid)] > 0 and
                                max_cnct_dwn[hnodes.index(mid)] > 0 and
                                l1.K_MAX[t] > 0 and
                                l2.K_MAX[t] > 0
                            ):
                                feasible_pairs.append((l1, l2))

                if feasible_pairs:
                    l1_sel, l2_sel = random.choice(feasible_pairs)
                    mid = l1_sel.n2

                    Z[links.index(l1_sel)][demands.index(d)][t] = 1
                    Z[links.index(l2_sel)][demands.index(d)][t] = 1

                    max_cnct_up[hnodes.index(mid)]  = 0
                    max_cnct_dwn[hnodes.index(mid)] = 0

        _, key_srv, a_lst = offline_lp(
            gnodes, hnodes, links, demands, 1, 2, Z
        )

        rnd_results_k_srv.append(key_srv * 1e-6)  # Mbps or Gbps normalization
        rnd_results_a_lst.append(a_lst   * 1e-6)  # Mbps or Gbps normalization

    # Statistics for plotting
    key_srv_rnd_qkp_avg[idx] = np.mean(rnd_results_k_srv)
    key_srv_rnd_qkp_min[idx] = np.min(rnd_results_k_srv)
    key_srv_rnd_qkp_max[idx] = np.max(rnd_results_k_srv)

    a_lst_rnd_qkp_avg[idx] = np.mean(rnd_results_a_lst)
    a_lst_rnd_qkp_min[idx] = np.min(rnd_results_a_lst)
    a_lst_rnd_qkp_max[idx] = np.max(rnd_results_a_lst)

    #########################################################################################################

    print(f"idx: {idx} | avg={key_srv_rnd_qkp_avg[idx]:.3f}")
    print(f"idx: {idx} | avg={a_lst_rnd_qkp_avg[idx]:.3f}")
```
