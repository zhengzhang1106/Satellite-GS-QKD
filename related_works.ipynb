```python
# -----------------------------
# Data extracted from the figure
# -----------------------------

# Fiber-based QKD points (blue X)
fiber_years = [2016, 2018, 2020, 2021, 2023]
fiber_distances = [60, 400, 500, 600, 1000]
fiber_rates = [
    "1.2 Mbps",
    "6.5 bps",
    "0.118 bps",
    "0.778 bps",
    "0.0034 bps"
]
#fiber_refs = ["[8]", "[4]", "[9]", "[17]", "[14]"]

# Free-space QKD point (red triangle)
free_years = [2017, 2025]
free_distances = [1200, 1705]
free_rates = ["1 Kbps", "2.975 Kbps"]
#free_refs = ["[13]", "[12]"]

# -----------------------------
# Plot
# -----------------------------

plt.figure(figsize=(9, 4))

# Fiber-based points
plt.scatter(
    fiber_years,
    fiber_distances,
    marker='x',
    s=150,
    linewidths=3,
    color='blue',
    label='Fiber-based'
)

# Free-space points
plt.scatter(
    free_years,
    free_distances,
    marker='^',
    s=180,
    color='red',
    label='Free-Space'
)

# -----------------------------
# Annotations
# -----------------------------

for x, y, rate in zip(fiber_years, fiber_distances, fiber_rates):
    plt.text(
        x,
        y + 40,
        f"{rate}",
        fontsize=16,
        ha='center',
        va='bottom'
    )

for x, y, rate in zip(free_years, free_distances, free_rates):
    plt.text(
        x,
        y + 40,
        f"{rate}",
        fontsize=18,
        ha='center',
        va='bottom'
    )

# -----------------------------
# Axes, grid, legend
# -----------------------------

plt.xlabel("Project Year", fontsize=16)
plt.ylabel("Achieved Distance (km)", fontsize=16)

plt.xlim(2015.1, 2025.9)
plt.ylim(0, 2100)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.grid()

plt.legend(fontsize=15, loc='lower right')

plt.tight_layout()

# Use only years that actually appear in the data
all_years = sorted(set(fiber_years + free_years))

plt.xticks(all_years, fontsize=16)
plt.savefig("intro.pdf", format='pdf', transparent=True, bbox_inches='tight', pad_inches=0.01)
plt.show()
```
