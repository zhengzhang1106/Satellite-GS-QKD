```python
gnodes, hnodes, links, demands = init_setup_real()
```

```python
# Initialize state and action
state  = [{} for _ in syst.T]
action = [{} for _ in syst.T]
reward = [0  for _ in syst.T]

state[0]["demands"] = demands
state[0]["a"] = {i: 0 for i in range(len(links))}
```

```python
# ---------- Hyperparams ----------
GAMMA = 0.95
LR = 5e-4
EPOCHS = 3
BATCH_SIZE = 64

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def build_state_vector_for_time(state_t, links, t, n_gnodes=len(gnodes), normalize=True):
    """
    Build state vector for the QKD environment.

    Includes:
    - QKP per link (normalized if normalize=True)
    - K_MAX per link (capacity)
    - For each demand: one-hot source, one-hot destination, normalized K_REQ

    State vector format:
    [ QKP_link0 .. QKP_link(L-1),
      KMAX_link0 .. KMAX_link(L-1),
      (src one-hot, dst one-hot, KREQ_demand0), ... ]
    """

    # -------------------------------
    # QKP / stored key bits per link
    # -------------------------------
    a_vec = np.array(
        [state_t.get("a", {}).get(idx_l, 0.0) for idx_l in range(len(links))],
        dtype=np.float32
    )

    if normalize:
        max_qkp = np.max(a_vec) if np.max(a_vec) > 0 else 1.0
        a_vec = a_vec / max_qkp

    # -----------------------------------------
    # Link Capacity (K_MAX per link)
    # -----------------------------------------
    if normalize:
        K_MAX_MAX = max(l.K_MAX[t] for l in links) + 1e-9  # avoid /0
        kmax_vec = np.array([l.K_MAX[t] / K_MAX_MAX for l in links], dtype=np.float32)
    else:
        kmax_vec = np.array([l.K_MAX[t] for l in links], dtype=np.float32)

    # ---------------------------
    # Encode demand information
    # ---------------------------
    K_REQ_MAX = max(d.K_REQ[t] for d in demands) + 1e-9

    d_vec = []

    # Loop over EVERY ordered GS pair (i → j, i ≠ j)
    for src in gnodes:
        for dst in gnodes:
            if src == dst:
                continue

            # Check whether there is a demand for this exact pair
            matched = next(
                (d for d in state_t.get("demands", [])
                 if d.n1 == src and d.n2 == dst),
                None
            )

            if matched:
                # Active demand exists
                kreq = float(matched.K_REQ[t])
                if normalize:
                    kreq = kreq / K_REQ_MAX

                src_idx = gnodes.index(src)
                dst_idx = gnodes.index(dst)

                src_norm = (src_idx + 1) / n_gnodes
                dst_norm = (dst_idx + 1) / n_gnodes

                d_vec.extend([src_norm, dst_norm, kreq])

            else:
                # No demand for this pair
                d_vec.extend([0.0, 0.0, 0.0])

    # ---------------------------
    # Concatenate full state
    # ---------------------------
    state_vector = np.concatenate(
        [a_vec, kmax_vec, np.array(d_vec, dtype=np.float32)]
    )

    return state_vector

class QNetwork(nn.Module):
    def __init__(self, state_dim, action_dim, hidden=256):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(state_dim, hidden),
            nn.ReLU(),
            nn.Linear(hidden, hidden),
            nn.ReLU(),
            nn.Linear(hidden, action_dim)
        )

    def forward(self, state):
        """
        state: [batch, state_dim]
        returns: Q-values [batch, action_dim]
        """
        return self.net(state)

class ReplayBuffer:
    def __init__(self, capacity=200_000):
        self.buffer = deque(maxlen=capacity)

    def push(self, s, a, r, s_next, done):
        self.buffer.append((s, a, r, s_next, done))

    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        s, a, r, s_next, done = zip(*batch)
        return (
            torch.stack(s),
            torch.stack(a),
            torch.tensor(r, dtype=torch.float32),
            torch.stack(s_next),
            torch.tensor(done, dtype=torch.float32)
        )

    def __len__(self):
        return len(self.buffer)

class DQNAgent:
    def __init__(self, state_dim, action_dim, lr=LR):
        self.q_net = QNetwork(state_dim, action_dim).to(DEVICE)
        self.target_q_net = QNetwork(state_dim, action_dim).to(DEVICE)
        self.target_q_net.load_state_dict(self.q_net.state_dict())
        self.target_q_net.eval()

        self.optimizer = optim.Adam(self.q_net.parameters(), lr=lr)
        self.buffer = ReplayBuffer()

        self.action_dim = action_dim
        self.epsilon = 1.0
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.999
        self.target_update_interval = 100
        self.step_count = 0

    def select_action(self, state_vec):
        if np.random.rand() < self.epsilon:
            return np.random.randint(0, 2, size=self.action_dim)

        with torch.no_grad():
            s = torch.tensor(state_vec, dtype=torch.float32, device=DEVICE).unsqueeze(0)
            q_vals = self.q_net(s)
            #print(f"q_vals: {q_vals}")
            return (q_vals > 0).int().squeeze(0).cpu().numpy()

    def update(self):
        if len(self.buffer) < BATCH_SIZE:
            return

        states, actions, rewards, next_states, dones = self.buffer.sample(BATCH_SIZE)

        states = states.to(DEVICE)
        actions = actions.to(DEVICE)
        rewards = rewards.to(DEVICE)
        next_states = next_states.to(DEVICE)
        dones = dones.to(DEVICE)

        # Q(s,a)
        q_vals = self.q_net(states)
        q_selected = (q_vals * actions).sum(dim=1)

        # max_a' Q_target(s', a')
        with torch.no_grad():
            next_q_vals = self.target_q_net(next_states)
            next_q_max = next_q_vals.clamp(min=0).sum(dim=1)

        target = rewards + GAMMA * next_q_max * (1 - dones)

        loss = F.mse_loss(q_selected, target)

        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

        # ε decay
        self.epsilon = max(self.epsilon * self.epsilon_decay, self.epsilon_min)

        # Target network update
        self.step_count += 1
        if self.step_count % self.target_update_interval == 0:
            self.target_q_net.load_state_dict(self.q_net.state_dict())
```

```python
# prepare agent
n_links = len(links)
n_demands = len(demands)   # assume constant across T
action_dim = n_links

# choose state_dim produced by build_state_vector_for_time
state_vector_example = build_state_vector_for_time(state[0], links, 0)
state_dim = len(state_vector_example)

# Training hyperparams
num_episodes = 5000
episode_len = len(syst.T)

# Track total rewards per episode
episode_rewards = []

solution = {"a": state[0]["a"].copy()}

agent = DQNAgent(state_dim, action_dim)

for ep in range(num_episodes):
    state = [{} for _ in syst.T]
    action = [{} for _ in syst.T]

    state[0]["demands"] = demands.copy()
    state[0]["a"] = {i: 0 for i in range(n_links)}

    total_reward = 0.0

    for t in syst.T:
        s_vec = build_state_vector_for_time(state[t], links, t)

        action_vec = agent.select_action(s_vec)

        action[t] = {idx_l: int(action_vec[idx_l]) for idx_l in range(n_links)}

        # --- environment step ---
        check_tx, check_rx = 1, 1
        sum_tx, sum_rx = {}, {}
        nodes = gnodes + hnodes

        for idx_n, n in enumerate(nodes):
            sum_tx[idx_n] = sum(action[t][idx_l] for idx_l, l in enumerate(links) if l.n1 == n)
            sum_rx[idx_n] = sum(action[t][idx_l] for idx_l, l in enumerate(links) if l.n2 == n)

            check_tx &= (sum_tx[idx_n] <= n.N_TX)
            check_rx &= (sum_rx[idx_n] <= n.N_RX)

        if check_tx and check_rx:
            solution, reward_t, A_next = online_new(gnodes, hnodes, links, state, action, t, 1)
        else:
            reward_t = -10

        total_reward += reward_t

        # next state
        if t + 1 < episode_len:
            next_state = state[t+1]
            next_state["a"] = A_next if check_tx and check_rx else state[t]["a"]
            next_state["demands"] = state[t]["demands"]
            s_next_vec = build_state_vector_for_time(next_state, links, t+1)
        else:
            s_next_vec = s_vec

        agent.buffer.push(
            torch.tensor(s_vec, dtype=torch.float32),
            torch.tensor(action_vec, dtype=torch.float32),
            reward_t,
            torch.tensor(s_next_vec, dtype=torch.float32),
            t == episode_len - 1
        )

        agent.update()

    episode_rewards.append(total_reward)
    print(f"Episode {ep+1}/{num_episodes}, reward={total_reward:.3f}, epsilon={agent.epsilon:.3f}")


# === Plot learning curve ===
plt.figure(figsize=(7,4))
plt.plot(range(1, num_episodes+1), episode_rewards, marker='o', markevery=num_episodes/10, linewidth=1.5)
plt.xlabel("Episode")
plt.ylabel("Total Reward over syst.T")
plt.title("Learning Progress of PPO Agent")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# === Plot learning curve with running average ===
window = 10                     # moving average window (change as desired)
running_avg = []

for i in range(len(episode_rewards)):
    start = max(0, i - window + 1)
    running_avg.append(np.mean(episode_rewards[start:i+1]))

plt.figure(figsize=(7,4))
plt.plot(range(1, num_episodes+1), running_avg, linewidth=2, label=f"Running Avg (window={window})")
plt.xlabel("Episode")
plt.ylabel("Smoothed Total Reward")
plt.title("Learning Progress of PPO Agent")
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
