```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family']  = 'DeJavu Serif'
plt.rcParams['font.serif']   = ['Times New Roman']
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype']  = 42

# ------------------------------------------------------------------
# Load flight data
# ------------------------------------------------------------------
df = pd.read_csv("flight_data.csv")

# Rename columns here if needed
# df = df.rename(columns={
#     "time_col_name": "Time",
#     "lat_col_name": "Latitude",
#     "lon_col_name": "Longitude",
#     "alt_col_name": "Altitude"
# })

# ------------------------------------------------------------------
# Extract data
# ------------------------------------------------------------------
# Extract hh:mm AM/PM from strings like "Sat 08:30:40 AM"
time_str = (
    df["Time"]
    .astype(str)
    .str.replace(r"^[A-Za-z]{3}\s+", "", regex=True)  # remove "Sat "
    .str.replace(r":\d{2}\s", " ", regex=True)        # remove seconds
)
lat = df["Latitude"].to_numpy()
lon = df["Longitude"].to_numpy() + 360 
alt_ft = (
    df["Altitude"]
    .str.replace(",", "", regex=False)  # remove thousands separator
    .astype(float)
)

alt_m = alt_ft.to_numpy() * 0.3048 * 1e-3

# Time index (used for plotting)
t_idx = np.arange(len(df))

# ------------------------------------------------------------------
# Plot 1: Trajectory (Latitude vs Longitude)
# ------------------------------------------------------------------
plt.figure(figsize=(3, 3))
plt.plot(lon, lat, marker="o", linewidth=2, markevery=len(df) - 1, color="red")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.axis("equal")

# Optional: annotate a few timestamps
step = max(1, len(df) - 1)
text = ["North Bay-YYB", "Barry's Bay"]
plt.text(lon[30], lat[0], text[0], fontsize=8)
plt.text(lon[len(df) - 90], lat[len(df) - 100], text[1], fontsize=8)
plt.savefig("flight_data_longlat.pdf", format='pdf', transparent=True, bbox_inches='tight', pad_inches=0.01)
plt.tight_layout()

# ------------------------------------------------------------------
# Plot 2: Altitude over Time
# ------------------------------------------------------------------
plt.figure(figsize=(6, 2))
plt.plot(t_idx, alt_m, marker="o", linewidth=2, markevery=len(df) - 1, color="red")
plt.xlabel("Time")
plt.ylabel("Altitude (km)")
plt.grid(True)
step = max(1, len(df)//6 - 1)
plt.xticks(t_idx[::step], time_str.iloc[::step], rotation=0)

plt.tight_layout()
plt.savefig("flight_data_alt.pdf", format='pdf', transparent=True, bbox_inches='tight', pad_inches=0.01)
plt.show()
```

```python
# ------------------------------------------------------------------
# Plot 3: 3D Trajectory (Longitude, Latitude, Altitude)
# ------------------------------------------------------------------
fig = plt.figure(figsize=(3, 4))
ax = fig.add_subplot(111, projection="3d")

ax.view_init(elev=15, azim=225)

ax.plot(
    lon,
    lat,
    alt_m,
    linewidth=1.5,
    color="red"
)

ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_zlabel("Altitude (km)")

# Optional: mark start and end points
ax.scatter(lon[0], lat[0], alt_m[0], s=60, marker="o", color="red", label="Start (North Bay's YYB)")
ax.scatter(lon[-1], lat[-1], alt_m[-1], s=60, marker="o", color="red", label="End (Near Barry's Bay)")
ax.scatter(lon[139], lat[139], alt_m[139], s=60, marker="o", color="red", label="End (Near Barry's Bay)")

# Optional: annotate a few timestamps
#step = max(1, len(df) - 1)
ax.text(lon[0]-0.1, lat[0]-0.1, alt_m[0], "North Bay-YYB\n8:30 AM", fontsize=10)
ax.text(lon[-1]+0.5, lat[-1]+0.5, alt_m[-1]-5, "Barry's Bay\n9:51 AM", fontsize=10)
ax.text(lon[139]-0.05, lat[139]-0.05, alt_m[139], "9:29 AM", fontsize=10)

#ax.legend()
plt.tight_layout()

plt.savefig("flight_data_3d.svg", format='svg', transparent=True, bbox_inches='tight', pad_inches=0.01)
plt.show()
```
